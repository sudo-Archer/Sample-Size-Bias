```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def newBaby():
    return np.random.randint(2) # boy is 1 and girl is 0
```

```python
def newDay(babyPerDay):
    return np.array([newBaby() for i in range(babyPerDay)])
```

```python
def boysMoreThan60(ListOfBabys):
    numOfboys = np.sum(ListOfBabys)
    if(numOfboys/len(ListOfBabys) >= 0.6):
        return 1
    else :
        return 0
    
```

```python
def getData(babyPerDay, numOfDays = 1000):
    counter = 0
    for i in range(numOfDays):
        counter += boysMoreThan60(newDay(babyPerDay))
    return counter
```

```python
hospital_A = getData(15)
hospital_A
```

```python
hospital_B = getData(45)
hospital_B
```

```python
yx = [getData(i) for i in range(1, 70)]
```

```python
plt.plot([i for i in range(1, 70)],yx)
plt.xlabel("Babys born per day")
plt.ylabel("Number of days that boys >= 60%")
```
